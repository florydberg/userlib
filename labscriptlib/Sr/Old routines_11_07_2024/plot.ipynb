```python
### plot
from scipy.optimize import curve_fit, least_squares
import numpy as np
import matplotlib.pyplot as plt

def linear(x, m, q):
    return q+m*x

fr=[77.9,77.67,77.02]
pow=np.array([20,14,7])



initial_guess = [0.1, 76.5]
low = [0, 0]
upper = [1, 80]
bounds = [low, upper]

# Fit the data using curve_fit
params, covariance = curve_fit(linear, pow, fr, p0=initial_guess, bounds=bounds )

# Extract the fitted parameters
mx, qx = params

plt.title('Stark Shift Profile')
plt.xlabel(f"Blue-Power(mW)\n m={(round(mx, 4))}, q={round(qx, 3)}")
plt.plot(pow, linear(pow, *params), color='red')
plt.plot(pow, fr, '--bo')
plt.legend(['Fitted'])
```
